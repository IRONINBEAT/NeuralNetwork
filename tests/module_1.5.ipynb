```python
from transformers import pipeline

def summarize_reviews(reviews):
    """
    Объединяет список отзывов и возвращает их краткое содержание.
    """
    # Объединяем 4 отзыва в один текст
    combined_text = " ".join(reviews)
    
    # Загружаем модель суммаризации
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
    
    # Получаем краткое содержание
    summary = summarizer(
        combined_text,
        max_length=150,
        min_length=40,
        do_sample=False
    )
    
    return summary[0]['summary_text']


if __name__ == "__main__":
    # Пример 4 отзывов
    reviews = [
        "Фильм впечатлил отличной игрой актеров и глубоким сюжетом. Особенно понравилась операторская работа.",
        "Картина получилась эмоциональной и напряженной. Музыка усиливала драматизм происходящего.",
        "Несмотря на затянутый финал, фильм оставил сильное впечатление и заставил задуматься.",
        "Отличная режиссура и продуманный сценарий делают этот фильм одним из лучших в жанре."
    ]
    
    result = summarize_reviews(reviews)
    
    print("Краткое резюме:")
    print(result)
```
