```python
from transformers import pipeline

# 1. Отзывы
reviews = [
    "Интерстеллар — это визуальный шедевр. Визуальные эффекты и саундтрек Ханса Циммера создают невероятную атмосферу космического путешествия.",
    "Сюжет фильма глубоко эмоционален. Отношения отца и дочери проходят через время и пространство, что заставляет сопереживать героям до слез.",
    "Научная составляющая, основанная на работах Кипа Торна, поражает. Черные дыры и сингулярность показаны максимально реалистично для кино.",
    "Это один из лучших фильмов Кристофера Нолана. Несмотря на трехчасовой хронометраж, картина смотрится на одном дыхании и дает пищу для размышлений."
]

# Объединяем отзывы в один текст
full_text = " ".join(reviews)

# 2. Инициализация модели суммаризации
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# 3. Получение краткого содержания
# max_length - максимальное количество слов в резюме
# min_length - минимальное количество слов
summary = summarizer(full_text, max_length=100, min_length=30, do_sample=False)

# 4. Вывод результата
print("--- Оригинальный объединенный текст ---")
print(full_text)
print("\n--- Краткое резюме (Summary) ---")
print(summary[0]['summary_text'])
```
