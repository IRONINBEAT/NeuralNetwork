Шаг 1: Используем легковесный образ Python

`FROM python:3.9-slim`

Шаг 2: Устанавливаем рабочую директорию внутри контейнера

`WORKDIR /app`

Шаг 3: Копируем файл зависимостей
Это делается отдельно, чтобы Docker кэшировал установку библиотек

`COPY requirements.txt .`

Шаг 4: Устанавливаем библиотеки без сохранения кэша pip

`RUN pip install --no-cache-dir -r requirements.txt`

Шаг 5: Копируем исходный код приложения и веса модели

`COPY . .`

Шаг 6: Открываем порт 8000 для FastAPI

`EXPOSE 8000`

Шаг 7: Запускаем приложение через сервер uvicorn

`CMD ["uvicorn", "main:app", "--host", "0.0.0.0", "--port", "8000"]`

```python
def create_dockerfile():
    dockerfile_content = """# Базовый образ
FROM python:3.9-slim

# Настройка окружения
WORKDIR /app
ENV PYTHONDONTWRITEBYTECODE 1
ENV PYTHONUNBUFFERED 1

# Установка зависимостей
COPY requirements.txt .
RUN pip install --no-cache-dir -r requirements.txt

# Копирование проекта
COPY . .

# Запуск
EXPOSE 8000
CMD ["uvicorn", "main:app", "--host", "0.0.0.0", "--port", "8000"]
"""
    
    with open("Dockerfile", "w", encoding="utf-8") as f:
        f.write(dockerfile_content)
    
    print("Файл Dockerfile успешно создан!")

if __name__ == "__main__":
    create_dockerfile()
```
